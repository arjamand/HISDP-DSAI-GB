## Arithmetic Operators

The Python interpreter can be used to express the arithmetic operators in a very simple way as described below:

```python
# For Addition
print(1 + 1) # Output: 2

# For Multiplication
print(2 * 3) # Output: 6

# For Integer Division
print(17 // 10) # Output: 1
print(-17 // 10) # Output: -2

# For Modulus
print(27 % 10) # Output: 7
print(-27 % 10) # Output: 3
```



````python
# Excercise 1
# For each of the following expressions, what value will the expression give? Verify your answers by typing the expressions into Python.

# a.9 - 3
# b.8 * 2.5
# c.9 / 2
# d.9 / -2
# e.9 // -2
# f.9 % 2
# g.9.0 % 2
# h.9 % 2.0
# i.9 % -2
# j.-9 % 2
# k.9 / -2.0


# Solution 
print(9 - 3)
print(8 * 2.5)
print(9 / 2)
print(9 / -2)
print(9 // -2)
print(9 % 2)
print(9.0 % 2)
print(9 % 2.0)
print(9 % -2)
print(-9 % 2)
print(9 / -2.0)
````
