#### Task 1: Define a function called hypotenuse that calculates the length of the hypotenuse of a right triangle when the other two sides are given. Use this function in a program to determine the length of the hypotenuse for each of the following triangles. The function should take two arguments of float type and return the hypotenuse as float type too. The sample output is as following:



````python
from math import sqrt
user_input = (str(input("Enter the sides of the triangle : "))).split()
base = float(user_input[0])
height = float(user_input[1])

def hypotenuse(side1, side2):
    print("Hypotenuse",sqrt((side1**2)+(side2**2)))

hypotenuse(base, height)

````

#### Task 2: Write a function distance that calculates the distance between two points (x1, y1) and (x2, y2). All numbers and return values should be of floating type. Use this function in your program. The sample output is as following:


````python
from math import sqrt
p1 = (str(input("Enter the first point : "))).split()
p2 = (str(input("Enter the second point : "))).split()
x1 = float(p1[0])
y1 = float(p1[1])
x2 = float(p2[0])
y2 = float(p2[1])
def distance(x1,x2,y1,y2):
    dresult = sqrt(((x2-x1)**2)+((y2-y1)**2))
    print(dresult)
    print(f"Distance between {x1,y1} and {x2,y2} is {dresult}")

distance(x1,x2,y1,y2)
````

#### Task 3: Write a function reversed that takes an integer value (between 1-9999) and returns the number with its digits reversed. For example, given the number 7631, the function should return 1367. Demonstrate the use of this function in your program with the following sample output.
```
Enter a number between 1 and 9999: 7631
The number with its digits reversed is: 1367


````python
def reversed(num = input("Enter a number between 1 and 9999 : ")):
    reversed_num = ""
    for i in range(1,(len(num)+1)) :
        reversed_num += num[-i]
    print("The number with its digits reversed is:",reversed_num)
reversed()
````

#### Task 4: Write a function is_prime that accepts an integer as argument and returns True if the number is prime and False otherwise. Use this function in your program to print on screen all the 4 digit prime numbers.

````python
def is_prime(num = int(input("Enter a integer to check for prime : "))):
    temp_list = []
    for i in range(2, num):
        if (num % i) == 0 :
            temp_list.append(i)
    if len(temp_list) == 0 :
        print("Given Number is a Prime .")
    else :
        print("Given Number is not a Prime .")
is_prime()
````

#### Task 6: Write a function is_bouncy that accepts an integer as argument and returns True if the number is bouncy and False otherwise. Use this function in your program to print on screen all the 4 digit bouncy numbers along with the total count of the 4-digit bouncy numbers.

````python
def is_bouncy(num = input("Enter a number to check is its bouncy : ")):
    list1 = [int(i) for i in num]
    list2 = sorted(list1)
    list3 = list2[::-1]

    if list1 == list2 or list1 == list3:
        print("Number is NOT Bouncy !")
    else:
        print("Number is Bouncy")

is_bouncy()
    
````

````python
from math import factorial

def binom(n, k):
    return ((factorial(n))) / ((factorial(k))*(factorial((n-k))))

print(binom(50,10))
````

#### Task 9: Write a program that plays the game of guess the number as follows: Your program chooses the number to be guessed by selecting an integer at random in the range 1 to 1000. The program then types:


#### The player then types a first guess. The program responds with one of the following:


#### If the players guess is incorrect, your program should loop until the player finally gets the number right. Your program should keep telling the player Too high or Too low to help the player zero in on the correct answer. The sample run of the game should be as following:



````python
from random import randint
def number_guesser():
    secret_number = randint(1,1000)
    user_input = int(input("I have a number between 1 and 1000.\nCan you guess my number?\nPlease type your first Guess : "))
    while user_input != secret_number :
        if user_input > secret_number :
            user_input = int(input("Too High.Try again : "))
        elif user_input < secret_number :
            user_input = int(input("Too Low.Try again : ")) 
    print("Excellent! You guessed the number!")

while True :
    number_guesser()
    play_again = int(input("Would you like to play again?\nPlease type ( 1-yes, 2-no ) : "))
    if play_again == 2 :
        break
    
    
````

````python
def perfect_num_checker(num = int(input("Enter a number : "))):
    checker = 0
    for i in range(1,num):
        if num%i == 0:
            checker += i
    if checker == num :
        return True
    else :
        return False

print("Number is a perfect Number :", perfect_num_checker())
        
    
````

#### Task 11 Write a function int2binary that accepts an integer as argument and returns its equivalent binary number using while loop and a list Use this function in a program that prints on screen the binary equivalent of numbers from 100 till 1000 in a tabular form.
[Hint: Use the division-by-2 algorithm]

````python
def int2binary(num = int(input("Enter number to convert into binary : "))):
    binary = []
    while True:
        binary.append(num%2)
        num = (num // 2)

        if num/2 == 0:
            break
    binary = binary[::-1]
    print(*binary)
int2binary()
            
````
