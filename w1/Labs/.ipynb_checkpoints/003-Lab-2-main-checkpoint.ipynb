#### Write a program to prompt for a score between 0.0 and 1.0. If the score is out of range, print an error message. If the score is between 0.0 and 1.0, print a grade using the following table:


````python
user_input = float(input("Enter Score between 0.0 - 1.0 for Grading : "))
if user_input <= 1.0 and user_input >= 0.0 :
    if user_input >= 0.9 :
        print("Grade : A")
    elif user_input >= 0.8 and user_input < 0.9 :
        print("Grade : B")
    elif user_input >= 0.7 and user_input < 0.8 :
        print("Grade : C")
    elif user_input >= 0.6 and user_input < 0.7 :
        print("Grade : D")
    elif user_input < 0.6 :
        print("Grade : F")
else :
    print("Error - Out of Range .")
````
